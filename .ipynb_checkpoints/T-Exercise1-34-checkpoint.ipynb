```python
df = pd.read_csv('data/Data.txt' , sep=" ",header=None)
df=pd.DataFrame(df)
df
```

```python
#df.iloc[0,0]
```

```python
# Remove the trailing semicolon
header=df.iloc[0,0].rstrip(";")
# Split the string into a list of values
header=header.split(";")
```

```python
# Create the temporary list containing all data without semicolon
# After this, data's shape''ll be 14643x13 (list dimension)
data=[] 
for i in range(df.shape[0]):#14644
    dataTmp=df.iloc[i,0].rstrip(";")
    dataTmp=dataTmp.split(";")
    data.append(dataTmp)   
```

```python
#Initial null dataframe
dfP=df = pd.DataFrame(np.nan, index=range(df.shape[0]), columns=data[0])
dfP
```

```python
# Transfer the data
for idx_r in range(1,df.shape[0]):
    for idx_ls in range(np.shape(data)[1]):
        dfP.iloc[idx_r-1,idx_ls]=data[idx_r][idx_ls]  
```
